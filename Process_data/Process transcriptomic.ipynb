```python
dp_rna = '/content/drive/MyDrive/Histopathology/Data/TCGA_KIRC/data_RNA_Seq_v2_mRNA_median_Zscores.txt'
df_rna = pd.read_csv(dp_rna, delimiter='\t').transpose()
df_rna.columns = df_rna.iloc[0]
df_rna = df_rna.drop(df_rna.index[0])
df_rna
```

```python
# Step 1: Drop columns where the header is NaN
df_rna = df_rna.loc[:, df_rna.columns.notna()]

# Step 2: Drop the first row
df_rna = df_rna.drop(df_rna.index[0])

# Step 1: Reset the index, converting 'Hugo_Symbol' to a column
df_rna = df_rna.reset_index()

# Step 2: Rename the new column from 'Hugo_Symbol' to 'case_id'
df_rna = df_rna.rename(columns={'index': 'case_id'})

df_rna
```

```python
df_groups = pd.read_csv('/content/drive/MyDrive/Histopathology/Data/TCGA_KIRC/signatures.csv')
df_groups
```

```python
def get_genes(gene_group, df_groups, df_meta):
    gene_list = df_groups[gene_group].dropna().tolist()
    all_gene = list(df_meta.columns)
    filtered_genes = []
    for gene_pattern in gene_list:
        # Construct pattern from the gene pattern
        #pattern = re.compile(gene_pattern+'_')
        pattern = re.compile(gene_pattern)

        # Filter all_gene based on the pattern
        matched_genes = [gene for gene in all_gene if pattern.match(gene)]
        #if len(matched_genes)>1:
            #print(matched_genes)
        filtered_genes.extend(matched_genes)
        df_filtered = df_meta.loc[:, ['case_id']+filtered_genes]
        df_filtered = df_filtered.drop_duplicates()

    print(len(filtered_genes))
    return df_filtered
```

```python
group_dfs ={}
for group in df_groups.columns:
    group_dfs[group] = get_genes(group, df_groups, df_rna)
```
